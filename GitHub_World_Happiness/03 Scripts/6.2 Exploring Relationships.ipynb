### 2.1 Creating correlation matrix 

```python
# Create a suibset sutable for exploring correlation between variables
subset_wh_cor=wh_merged[['happiness_rank', 'happiness_score', 'economy_GDP_per_capita', 'family', 'health_life_expectancy', 'freedom', 'trust_government_corruption', 'generosity']]
```

### 2.2 Creating a correlation heatmap

```python
plt.matshow(subset_wh_cor.corr())
plt.show()
```

```python
# Add labels, a legend, and change the size of the heatmap
f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(subset_wh_cor.corr(), fignum=f.number) # type of plot
plt.xticks(range(subset_wh_cor.shape[1]), subset_wh_cor.columns, fontsize=14, rotation=90) # x axis labels
plt.yticks(range(subset_wh_cor.shape[1]), subset_wh_cor.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

```python
# Create a correlation heatmap using seaborn
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr=sns.heatmap(subset_wh_cor.corr(), annot=True, ax=ax) # The `annot` argument allows the plot to place the correlation coefficients onto the heatmap.
```

#### 1-	The correlation matrix shows a strong correlation between happiness rank and happiness score. The heatmap created in seaborn indicated a strong negative correlation. That is what we can expect, the highest happiness score is assigned to the lowest happiness rank. As far as the happiness score decreases, the country is going down the rank ladder. 

#### 2- Between the happiness score and the six happiness indicators, the highest level of correlation is with GDP per capita. The reason may be that GDP is a widely used indicator, it can be calculated using an accurate formula, and it is the least subjective among the six indicators used to rank happiness.

#### 3- Besides GDP per capita, a strong correlation is noted between the happiness score and health and life expectancy – 0.68, family (or social relationship) – 0.65, freedom of choice 0.57. Such indicators as social relationship and freedom of choice are subjective factors, nevertheless, when answering the survey questions, the correspondents are more or less objective as this is related to their personal experience.

#### 4- The correlation coefficient between the happiness score and two other indicators (perception of corruption and generosity) witnesses a week relationship. When answering to the question about corruption, the correspondents may rely mostly not on their experience, but the general perception existing in the society and disseminated by media.  

#### 5- All the six happiness indicators have positive correlation with the happiness score. As long as an indicator is going up, the people feel happier. I expected that on the contrary, there would be a negative correlation between happiness and the level of corruption, e.g. the higher corruption, less happy people are. This should be clarified further.

### 2.5 Creating a Categorical Plot

```python
# Create a histogram of happiness_score 
sns.histplot(subset_wh_cor['happiness_score'], bins=20, kde=True)
```

```python
# Split a categorical variable 'happiness_score' into happiness categories
wh_merged.loc[wh_merged['happiness_score']<4, 'happiness_category']='low'
```

```python
wh_merged.loc[(wh_merged['happiness_score']>=4) & (wh_merged['happiness_score']<5.5), 'happiness_category']='middle-low'
```

```python
wh_merged.loc[(wh_merged['happiness_score']>=5.5) & (wh_merged['happiness_score']<7), 'happiness_category']='middle-high'
```

```python
wh_merged.loc[wh_merged['happiness_score']>=7, 'happiness_category']='high'
```

```python
wh_merged['happiness_category'].value_counts(dropna=False)
```

```python
wh_merged.columns
```

```python
# Create a subset of 5 most populated countries
top_countries=['China', 'India', 'Indonesia', 'Pakistan', 'United States']
subset_wh_5_countries=wh_merged.loc[wh_merged['country'].isin(top_countries)]
```

```python
subset_wh_5_countries.head()
```

```python
subset_wh_5_countries.shape
```

```python
subset_wh_5_countries.info()
```

```python
# Create a categorical plot in seaborn using the happiness categories and Family factor
sns.set(style="ticks")
cat_plot = sns.catplot(x="country", y="family", hue="happiness_category", height=4, aspect=2, data=subset_wh_5_countries)
plt.xticks(rotation=90)
```

```python
# Create a categorical plot in seaborn using the happiness categories and Life expectancy factor
sns.set(style="ticks")
cat_plot=sns.catplot(x="country", y="health_life_expectancy", hue="happiness_category", height=4, aspect=2, data=subset_wh_5_countries)
plt.xticks(rotation=90)
```

```python
# Create a categorical plot in seaborn using the happiness categories and generosity factor
sns.set(style="ticks")
cat_plot=sns.catplot(x="country", y="generosity", hue="happiness_category", height=4, aspect=2, data=subset_wh_5_countries)
plt.xticks(rotation=90)
```

#### The happiness indicators don’t impact the level of happiness in the same way. For instance, the people with high-level happiness have not received the highest level of family support. While in India, the low happiness associated with lower level of family support. I assume the impact of cultural differences in western and Asia countries. 

#### Higher level of life expectancy is associated with high happiness level in United States. However, in India the people who lives longer life don’t feel happier. 

#### In the United States, people, who have reported the highest happiness level, witnesses a high generosity contribution. Among the top-5 populated countries, the highest rank of generosity is recorded in Indonesia. While in China the level of how much people donate is very low regardless that the fact that China is in top of countries by GDP.
