## 02. Merging dataframes

```python
# Create the list of dataframes to merge based on name convention
file_list = [path + f for f in os.listdir(path) if f.startswith('wh_20')]
```

```python
print(file_list)
```

```python
# Create an empty list to include the content of each dataframe converted to pandas DF
csv_list=[]
```

```python
# Read each (sorted) dataframe in file_list, convert it to pandas DF and appends it to the csv_list
for file in sorted(file_list):
    csv_list.append(pd.read_csv(file).assign(File_Name = os.path.basename(file)))
```

```python
# Merge the pandas DFs into one DF, index is refreshed 
csv_merged=pd.concat(csv_list, ignore_index=True)
```

```python
# The single DF to be saved to the path in CSV format, without index column
csv_merged.to_csv(path+'wh_merged.csv', index=False)
```

## 04. Missing values

```python
wh_merged_1.isnull().sum()
```

```python
wh_merged_1_nan
```

```python
# Define the mean value of health_life_expectancy in Palestine
palestine_data=wh_merged_1[wh_merged_1['country']=='Palestine']
mean_health_life_expectancy=palestine_data['health_life_expectancy'].mean()
print(mean_health_life_expectancy)
```
